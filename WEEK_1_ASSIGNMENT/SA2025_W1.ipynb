### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['hp_weight_ratio'] = df['horsepower'] / df['weight']
median_mpg = df['mpg'].median()
above_median_mpg = df[df['mpg'] > median_mpg]

max_ratio = above_median_mpg['hp_weight_ratio'].max()
best_ratio_cars = above_median_mpg[above_median_mpg['hp_weight_ratio'] == max_ratio]

best_ratio_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_weight_ratio']]
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
grouped_data = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set plot style
sns.set(style="whitegrid")

# Create a multi-line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped_data, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()

plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
year_counts = df.groupby('name')['model_year'].nunique()

multi_year_cars = year_counts[year_counts > 1].index
filtered_df = df[df['name'].isin(multi_year_cars)]

consistency_stats = (
    filtered_df.groupby('name')
    .agg(
        mpg_std=('mpg', 'std'),
        mpg_mean=('mpg', 'mean'),
        appearances=('mpg', 'count')
    )
    .reset_index()
)

consistent_models = consistency_stats[consistency_stats['mpg_std'] < 1.0]
sorted_consistent_models = consistent_models.sort_values(
    by=['appearances', 'mpg_mean'], ascending=[False, False]
)

sorted_consistent_models[['name', 'appearances', 'mpg_mean']]
```
